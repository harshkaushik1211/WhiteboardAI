```python
import pathlib

for f in pathlib.Path("/content/SadTalker").rglob("*.py"):
    txt = f.read_text(errors="ignore")

    txt = txt.replace("astype(float32)", "astype(np.float32)")
    txt = txt.replace("dtype=float32", "dtype=np.float32")
    txt = txt.replace(" img.astype(float32)", " img.astype(np.float32)")

    f.write_text(txt)

print("patched")
```
